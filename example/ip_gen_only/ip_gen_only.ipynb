# IP Generation Only

This notebook demonstrates how to compose and package all DFX4ML IP cores
**without** running synthesis or implementation.

Use this when you only need to populate `ip_repo/` (e.g. to inspect packaged IPs,
re-generate after modifying IP source, or prepare the repo for a manual Vivado flow).

The following IPs will be packaged into `<build_folder>/ip_repo/`:
- `dfx_icap`
- `dfx_mng`
- `dfx_streamer`
- `dfx_streamer_mreset`
- `dfx_streamer_mshut`
- `dfx_streamer_s2m`
- `dfx_streamer_s2m_free`
- `dfx_streamer_sreset`
- `dfx_streamer_sshut`

```python
import sys
sys.path.insert(0, "../..")

from lib.hw_build import HwBuildHelper
```

## Parameters

Only three parameters are required for IP-only generation:

| Parameter | Description |
|---|---|
| `build_folder_path` | Output folder where `ip_repo/` will be created |
| `dfx_root_path` | Root of the DFX4ML repository |
| `vivado_path` | Path to the Vivado executable |

```python
build_folder_path = "./build_ip_only"
dfx_root_path     = "../.."  # adjust if running from a different location
vivado_path       = "/tools/Xilinx/Vivado/2023.2/bin/vivado"
```

## Create helper and check mode

```python
hw_builder = HwBuildHelper.for_ip_only(
    build_folder_path = build_folder_path,
    dfx_root_path     = dfx_root_path,
    vivado_path       = vivado_path,
)

print(f"ip_only_mode : {hw_builder.ip_only_mode}")
print(f"build folder : {hw_builder.build_folder_path}")
print(f"DFX4ML root  : {hw_builder.dfx_root_path}")
```

## Run IP composition

Launches Vivado to package all DFX4ML IPs into `<build_folder>/ip_repo/`.
Vivado will open in GUI mode; it exits automatically when packaging is complete.

```python
hw_builder.build_ip_only()
```
