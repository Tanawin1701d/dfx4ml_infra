```python
from lib.hw_build import HwBuildHelper
from lib.sw_build import SwBuildHelper
```

```python
# Single board_build_tcl shared across all iterations
board_build_tcl = "./kv260/board_build.tcl"

# One constraint XDC per iteration
constraint_xdc_list = [
    "./kv260/constraint_1.xdc",
    "./kv260/constraint_2.xdc",
]

# Common parameters shared across all iterations
common_params = dict(
    dfx_root_path="../..",
    user_repo_path="",
    user_rm_build_tcl_path="",
    req_gen_ip=1,
    num_core=4,
    clk_frq=99999001,  # Hz
    rm_index_width=2,
    num_dfx_streamer=2,
    interface_widths=[32, 32],
    applied_interface_widths=[32, 32],
    storage_index_widths=[10, 10],
    num_actual_rm=2,
    input_map_list=[[0, -1], [-1, 0]],
    output_map_list=[[-1, 0], [0, -1]],
    ip_map_list=["", ""],
    test_mode=1,
    vivado_path="/tools/Xilinx/Vivado/2023.2/bin/vivado",
)
```

```python
for i, constraint_xdc in enumerate(constraint_xdc_list):
    print(f"=== Build iteration {i} ===")
    hw_builder = HwBuildHelper(
        build_folder_path=f"./build_prj_{i}",
        export_folder_path=f"./export_{i}",
        board="custom",
        board_build_tcl=board_build_tcl,
        constraint_xdc=constraint_xdc,
        **common_params,
    )
#    hw_builder.run_build()
    hw_builder.package_export_files()

    sw_builder = SwBuildHelper(export_folder_path=f"./export_{i}")
    sw_builder.package_export_file()
```
